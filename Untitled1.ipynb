```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random
import math
import time
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
import datetime
import operator
plt.style.use('seaborn')
%matplotlib inline
```

```python
df_geral = pd.read_csv('data/basegeral.csv', sep=';')
```

```python
df_geral['Sexo'] = df_geral['Sexo'].str.upper()
df_geral.head()
```

```python
grouped = df_geral.groupby('dt_notificacao')
```

```python
lista_homem = []
lista_mulher = []
lista_indefinido = []
lista_data = []
for name, group in grouped:
    lista_mulher.append(len(group[group['Sexo']=="FEMININO"]))
    lista_homem.append(len(group[group['Sexo']=="MASCULINO"]))
    lista_indefinido.append(len(group[group['Sexo']=="INDEFINIDO"]))
    lista_data.append(name)
```

```python
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=lista_data,
        y=lista_mulher,
        name="Sexo Feminino"
    ))

fig.add_trace(
    go.Scatter(
        x=lista_data,
        y=lista_homem,
        name="Sexo Masculino"
    ))
fig.add_trace(
    go.Scatter(
        x=lista_data,
        y=lista_indefinido,
        name="Sexo Indefinido"
    ))

py.iplot(fig)
```

### Anlise do Grfico

Como podemos ver a relao de casos entre homens e mulheres  muito parecida e sempre uma queda entre os picos. Essa queda sugere uma subnotificao em alguns dias da semana como j  sabido que ocorre. Mas podemos verificar isso com um simples experimento.

Iremos assumir que isso  verdade para os prximos experimentos. 

Todos experimentos preditivos usaro o modelo do facebook (FbProphet) com taxa de confiana de 95%

```python
df_sintetico2 = df_sintetico.copy()
df_sintetico2 = df_sintetico2.drop('dt_atualizacao', axis=1)
df_sintetico2 = df_sintetico2.add_prefix('y')
df_sintetico2 = df_sintetico2.rename({'ydt_referencia':'ds'}, axis=1)
```

```python
df_test = df_sintetico2[['ds', 'yconfirmados']]
df_test = df_test.rename({'yconfirmados':'y'}, axis=1)
```

```python
from fbprophet import Prophet
```

```python
model = Prophet(interval_width=0.95)
model.fit(df_test)
```

```python
future_dates = model.make_future_dataframe(periods=12, freq='MS')
```

```python
forecast = model.predict(future_dates)
```

```python
model.plot(forecast,
              uncertainty=True)
```

### Anlise
Como podemos ver, num cenrio simples onde no possumos os dados de queda da curva ou do total da populao (geral e percentual contaminado), no podemos dizer que a curva em algum momenento vai descer. No momento a previso para os prximos 12 meses  s a subida do nmero de casos confirmados.

```python
model.plot_components(forecast)
```

### Anlise

Utilizando o total de casos confirmados  notrio que o nmero de caso notificados chega a aumentar no meio da semana e ser menor na segunda e finais de semana pela subnotificao do nmero de casos como predito no cenrio anterior.
